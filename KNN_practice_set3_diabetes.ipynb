```python
import numpy as np #numerical computations
from collections import Counter #count the values
```

```python
# We have some Features of Diabetes [Pregnancies, Glucose, BloodPressure, BMI, Age]
X_train = np.array([   #load old dataset
    [2, 120, 70, 25.3, 31],
    [0, 95, 60, 22.0, 25],
    [3, 130, 80, 28.1, 35],
    [1, 85, 65, 20.0, 22],
    [5, 160, 90, 32.0, 45]
])

# Labels: 1 = Diabetes, 0 = No Diabetes
y_train = np.array([1, 0, 1, 0, 1])
```

```python
X_test=np.array([3, 150, 85, 29.0, 38]) #new data points
k=3 #assign the k value
```

```python
distances=[] #empty list
for i in range(len(X_train)):
  distance=np.sqrt(np.sum(X_train[i]-X_test)**2) #distance formula
  distances.append((distance,y_train[i])) #append:add values
```

```python
distances.sort(key=lambda x:x[0]) #sorting value
```

```python
k_nearest_labels=[label for distance,label in distances[:k]]
```

```python
prediction=Counter(k_nearest_labels).most_common(1)[0][0] #predict values
```

```python
if prediction==1:
  print("patient has diabetes")
else:
  print("no diabetes")
```
